```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsOneClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
```

```python
!pip3 install -U ucimlrepo
from ucimlrepo import fetch_ucirepo

# fetch dataset
census_income = fetch_ucirepo(id=20)
X = census_income.data.features
y = census_income.data.targets

y['income'] = y['income'].str.rstrip('.')

y
```

```python
label_encoder = []

for column in X.columns:
  if X[column].dtype == 'object':
    encoder = preprocessing.LabelEncoder()
    X[column] = encoder.fit_transform(X[column])
    label_encoder.append(encoder)

X
```

```python
classifier = OneVsOneClassifier(LinearSVC(random_state=0))
classifier.fit(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
classifier = OneVsOneClassifier(LinearSVC(random_state=0))
classifier.fit(X_train, y_train)
y_test_pred = classifier.predict(X_test)
f1 = cross_val_score(classifier, X, y, scoring='f1_weighted', cv=3)
print("F1 score: " + str(round(100*f1.mean(), 2)) + "%")
report = classification_report(y_test, y_test_pred)
print(report)
```

```python
input_data = [63, 'Local-gov', 65479, '9th', 5, 'Married-civ-spouse', 'Transport-moving', 'Husband', 'White', 'Male', 7688, 0, 40, 'United-States']

print(len(label_encoder))

# Кодуємо input_data за допомогою label_encoder
encoded_input_data = []
count = 0
for i, value in enumerate(input_data):
    if isinstance(value, str):  # Перевіряємо, чи є значення рядком
        encoded_input_data.append(label_encoder[count].transform([value])[0])
        count = count + 1
    else:
        encoded_input_data.append(int(value))

print(encoded_input_data)
```

```python
encoded_input_data = np.reshape(encoded_input_data, (1, 14))
test_data = pd.DataFrame(encoded_input_data)
predicted_class = classifier.predict(test_data)
predicted_class
```
